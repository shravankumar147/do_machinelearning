```python
def vpercp(train, target, weights, nb_epoch=1,kfold=10):
    w1 = weights
    X = train
    y = target
    c1 = 1

    v1 = []
    p1 = []
    
    for epoch in range(nb_epoch):
        print("*************************")
        print("Epoch {}".format(epoch))
        print("*************************")
        print(" ")
        print("current weights {}".format(w1))
        print(" ")
        for i in range(len(data)):
            print(" ")
            print("==>iteration{} on X{}".format(i,i))
            print(" ")
            u = np.dot(X[i],w1)
    #         print("{}-{}".format(u,y[i]))
            g = y[i]*u
            if g<=0:
                print(" ")
                print("misclassified")
                print(" ")
                w1 = w1+y[i]*X[i]
                v1.append(w1)
                p1.append(c1)
                c1 = 1
                print("Vote for the weight {}".format(c1))
            else:
                c1 =c1+1
                
            print(" ")
            print("classified")
            print(" ")
            print("Vote for the weight {}".format(c1))
    return v1,p1            
```

```python
def sign(x):
    if x<0:
        return -1
    return 1

def  vp_prediction(v, p, X):
    predictions = []
    for i in range(0,X.shape[0]):
        s  = 0
        for k in range(0, len(v)):
            s = s + p[k]*sign(np.dot(v[k],X[i,:]))
        predictions.append(sign(s))
    return predictions

def p_prediction(v, X):
    predictions = []
    for i in range(0, X.shape[0]):
        predictions.append(sign(np.dot(v[-1], X[i,:])))
    return predictions
def accuracy(predictions, labels):
    return np.sum((predictions==labels)).astype('float')/len(predictions)    
```
